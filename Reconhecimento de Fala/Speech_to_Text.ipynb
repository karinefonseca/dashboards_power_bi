```
#audio_rec = sr.Recognizer()
with sr.AudioFile('Merry.wav') as source:
  #use the Merry Christmas-SoundBible.com-1120316507.wav as the AudiFIle - which is our source

  #listen for the first phrase and extract it into audo data
  speech= audio_rec.listen(source)

  try: 
  #using google sppech recognition
    text = audio_rec.recognize_google(speech)
    print('You Said: ')
    print(text)

  except:
    print('could not undertand audio')
```
